**Таблицы №5-10, 13-16**

Таблицы №5, 6, 7, 8, 9, 10, 13, 14, 15, 16 идентичны по своей структуре и имеют единый алгоритм обработки

```python
onko = [5, 6, 7, 8, 9, 10, 13, 14, 15, 16]
onko_tab = [df_5, df_6, df_7, df_8, df_9, df_10, df_13, df_14, df_15, df_16]
```

```python
tab_result = []
```

```python
# Цикл для обработки таблиц
for i in onko_tab:
    i.columns = ['Локализация, нозологическая форма', 'Код МКБ 10', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
    i = i.dropna().reset_index(drop=True)
    i['mkb_10+local'] = i['Код МКБ 10'] + ' ' + i['Локализация, нозологическая форма']
    i.drop(['Локализация, нозологическая форма', 'Код МКБ 10'], axis= 1 , inplace= True )
    i = i.set_index('mkb_10+local')
    i = i.T.reset_index()
    i = i.replace(regex={',': '.'})
    i = i.replace(regex={'-': 0})
    i.rename(columns = {'index':'year'}, inplace = True)
    i = pd.melt(i, id_vars=['year'], var_name='mkb_10+local', value_name='value')
    i['year'] = i['year'].astype('int')
    i['value'] = i['value'].astype('float')
    i['mkb_10'] = i['mkb_10+local'].apply(lambda x: x.split(" ", 1)[0])
    tab_result.append(i)
```

```python
# Цикл для сохранения отредактированных таблиц

for n in range(len(onko)):   
    exec(f"df_{onko[n]} = tab_result[n]")
```

```python
# Цикл для вывода результата обработки таблиц

for n in range(len(onko)):
    print()
    print(f'Таблица №{onko[n]}')
    display(tab_result[n].head())
    print()
    tab_result[n].info()
```

**Таблица №17**

Переименуем столбцы:  
`year` - Годы  
`first_abs` - Абсолютное число больных с впервые в жизни установленным диагнозом 
  
Из числа больных с впервые в жизни установленным диагнозом:  
  `first_act` - выявлено активно, %  
  `first_morf` - диагноз подтвержден морфологически, %  
    
  имели стадию заболевания, %:  
    `first_stad_1_2` - I-II  
    `first_stad_3` - III  
    `first_stad_4` - IV  
    `first_stad_na` - не установлена  
      
`let_first_year` - Летальность на первом году с момента установления диагноза, %  
`let` - Летальность, %  
  
Находились под наблюдением на конец года:  
  `nabl_abs` - абс. число  
  `nabl_100` - на 100 тыс. детск. населения  
    
   из них 5 лет и более:  
      `nabl_100_5_abs` - абс. число  
      `nabl_100_5_%` - %  
        
`ind_nakopl` - Индекс накопления контингентов
