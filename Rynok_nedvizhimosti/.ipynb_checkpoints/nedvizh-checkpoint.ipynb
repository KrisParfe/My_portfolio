### Обработка пропущенных значений

```python
data.isna().sum()
```

Первый столбец с пропущенными значениями **`'ceiling_height'`** - высота потолков в метрах. Всего 9195 пропусков. Посмотрим на несколько строк из таблицы с пропущенными значениями высоты потолков на предмет наличия каких-либо закономерностей.

```python
data[data['ceiling_height'].isna()].head()
```

```python
data['ceiling_height'].value_counts().head(10)
```

Каких-либо закономерностей в появлении пропусков в столбце не выявлено. При этом пропуски составляют почти 40% от датасета и просто удалить их будет некорректно. Стандартная высота потолков в квартирах колеблется в небольшом диапазоне, в нашем наборе данных большинство квартир имеет высоту потолков от 2,5 до 3,2 метров, поэтому заменим пропущенные значения на медиану.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

```python
data['ceiling_height'].isna().sum()
```

Следующий столбец с пропусками значений **`'floors_total'`** — всего этажей в доме. Пропущено 86 значений.

```python
data['floors_total'].isna().sum()
```

```python
data[data['floors_total'].isna()].head()
```

```python
data[data['floors_total'].isna()]['floor'].value_counts()
```

```python
data['floors_total'].mean()
```

Какой-то взаимосвязи между пропущенными значениями в столбце 'floors_total' и значениями в других столбцах не выявлено. Наиболее вероятным было предположение, что пропуски как-то связаны со значениями в столбце 'floor', но и это предположение не подтвердилось. Заменить пропуски средним или медианой будет некорректно, т.к. в некоторых объявлениях этаж квартиры будет превышать общую этажность дома. Пропуски составляют всего 0,3% от выборки, поэтому удалим эти строки.

```python
data = data.dropna(subset=['floors_total'])
```

```python
data['floors_total'].isna().sum()
```

Следующий столбец с пропусками **`'living_area'`** — жилая площадь в квадратных метрах (м²). Всего 1870 пропусков.

```python
data['living_area'].isna().sum()
```

```python
data[data['living_area'].isna()].head()
```

```python
data[data['living_area'].isna()]['total_area'].value_counts()
```

Никакой закономерности в пропусках не обнаружено. 
У нас полностью заполнен столбец с общей площадью квартиры. Рассчитаем медианную долю жилой площади по тем строкам, где значения не пропущены, и заполним пропуски в жилой площади умножив соответствующую общую площадь на рассчитанную долю.

```python
data['living_area'] = (
    round(data['living_area'].fillna(data['total_area'] * (data['living_area'] / data['total_area']).median()), 2)
)
```

```python
data['living_area'].isna().sum()
```

Следующий столбец с пропусками **`'is_apartment'`** — апартаменты.

```python
data['is_apartment'].isna().sum()
```

```python
data['is_apartment'].unique()
```

```python
data[data['is_apartment'] == True].sample(5)
```

```python
data[data['is_apartment'] == False].sample(5)
```

Судя по объявлениям на Яндекс недвижимости, апартаменты продаются достаточно редко, их доля около 3% от общего числа объявлений  (по данным на сентябрь 2022 года). Поэтому сделаем предположение, что пропущенные значения означают False.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Следующий столбец с пропусками - **`'kitchen_area'`**

```python
data['kitchen_area'].isna().sum()
```

Предположим, что размер кухни в большей степени зависит от количества комнат, т.к. чем больше комнат, тем большее количество людей проживает, тем больше нужна кухня для размещения всех жильцов.

```python
data.pivot_table(index='rooms', values='kitchen_area', aggfunc='median')
```

```python
for t in data['rooms'].unique():
    data.loc[(data['rooms'] == t) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['rooms'] == t), 'kitchen_area'].median()
```

```python
data['kitchen_area'].isna().sum()
```

```python
data[data['kitchen_area'].isna()].sample(5)
```

```python
data[data['rooms'] == 0]['total_images'].count()
#посчитаем количество квартир с 0 комнат
```

```python
data[data['kitchen_area'].isna()]['studio'].value_counts()
```

После замены пропусков площади кухни на медианные значения по количеству комнат остались не заполнены строки с количеством комнат равных 0. Это студии, либо квартиры свободной планироки, в которых нет отдельных жилых комнат и отдельной кухни как таковой. Поэтому заполним оставшиеся пропуски нулями и убедимся, что пропусков не осталось.

```python
data['kitchen_area'] = data['kitchen_area'].fillna(0)
```

```python
data['kitchen_area'].isna().sum()
```

### Обработка неявных дубликатов

```python
data['locality_name'].unique()
```

```python
len(data['locality_name'].unique())
```

Выявлены дубликаты, возникающие из-за чередования е/ё, а также разного написания типа населенного пункта ('поселок городского типа' / 'городской поселок', 'поселок станции' / 'поселок при железнодорожной станции'), также есть несколько населенных пунктов с одинаковым названием, но разными типами, которые при поиске информации на Яндекс картах оказались одни и тем же населенным пунктом ('деревня Калитино' / 'поселок Калитино', 'деревня Кудрово' / 'Кудрово', 'поселок Мурино' / 'Мурино', 'деревня Федоровское' / 'городской поселок Федоровское', 'поселок Гарболово' / 'деревня Гарболово').

```python
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
```

```python
def replace_wrong_values(wrong_values, correct_value):  
    data['locality_name'] = data['locality_name'].replace(wrong_values, correct_value)
```

```python
data['locality_name'] = data['locality_name'].str.replace('поселок городского типа', 'городской поселок')
```

```python
data['locality_name'] = data['locality_name'].str.replace('поселок станции', 'поселок при железнодорожной станции')
```

```python
replace_wrong_values('деревня Калитино', 'поселок Калитино')
```

```python
replace_wrong_values('деревня Кудрово', 'Кудрово')
```

```python
replace_wrong_values('поселок Мурино', 'Мурино')
```

```python
replace_wrong_values('деревня Федоровское', 'городской поселок Федоровское')
```

```python
replace_wrong_values('поселок Гарболово', 'деревня Гарболово')
```

```python
data['locality_name'] = data['locality_name'].str.lower()
```

```python
data['locality_name'].unique()
```

```python
len(data['locality_name'].unique())
```

<div style="border:solid orange 3px; padding: 20px"> Помимо вышеперечисленных преобразований также привели все названия к строчному регистру, чтобы исключить возможные пропуски из-за разного написания названий с заглавной или строчной буквы.
После замены всех дубликатов количество уникальных населенных пунктов уменьшилось на 48 единиц.

```python
data['ceiling_height'].unique()
```

В наборе содержатся аномальные значения, есть высота потолков 32, 24, 27, и даже 100 метров. Т.к. эти данные заполнялись вручную, то скорее всего была допущена ошибка при заполнении и для некоторых значений точка была поставлена не в том месте. Заменим такие значения. Т.к. стандартная высота потолков около 2-3 метров, то возьмем все значения больше 20.

```python
data.loc[(data['ceiling_height'] >= 20), 'ceiling_height'] = data.loc[(data['ceiling_height'] >= 20), 'ceiling_height'] / 10
```

```python
data['ceiling_height'].unique()
```

Замена некорректно заполненных значений прошла успешно, но ещё остались очень высокие (более 5 метров) и очень низкие (менее 2 метров) значения. Посмотрим на них.

```python
data[data['ceiling_height'] >=5]
```

```python
data[data['ceiling_height'] < 2]
```

Квартиры с потолками 5-6 метров имеют место быть с учетом их площади и стоимости, встречаются объявления о продаже элитного жилья с очень высокими потолками, а 8 метров и выше в данной выборке имеют неправдоподобно маленькую площадь и низкую стоимость. Таким образом удалим строки с высотой потолков ниже двух и больше 8 метров

```python
data = data[(data['ceiling_height'] >= 2) & (data['ceiling_height'] < 8)]
```

### Средняя цена кв.м. в 10 населенных пунктах с наибольшим числом объявлений

```python
data_top_10 = (
    data.pivot_table(index='locality_name', values='last_price', aggfunc='count')
    .sort_values(by='last_price', ascending=False).head(10)
)
data_top_10
#создаем сводную таблицу с топ-10 населенных пунктов по количеству продаж квартир
```

```python
data_top_10_2 = data.pivot_table(index='locality_name', values='price_m2')
#создаем вторую сводную таблицу, в которой будет рассчитана средняя стоимость 1 кв.м. для каждого населенного пункта
```

```python
data_top_10 = data_top_10.join(data_top_10_2)
#присоединяем к первой таблице с топ-10 таблицу с данными о средней стоимости 1 кв.м.
```

```python
data_top_10.columns = ['count', 'mean_price_m2'] 
#переименовываем столбцы
```

```python
data_top_10['mean_price_m2'] = round(data_top_10['mean_price_m2'], 2)
data_top_10.sort_values(by='mean_price_m2', ascending=False)
#округляем данные до сотых и сортируем по убыванию стоимости
```

<div style="border:solid orange 3px; padding: 20px"> Самая высокая цена за 1 кв.м. среди топ-10 по проданным квартирам в Санкт-Петербурге - 114,3 тыс. рублей, на втором месте - Пушкин и 103,1 тыс. рублей за квадратный метр. Замыкает тройку Кудрово - 95,3 за 1 кв.м. Самый дешевый квадратный метр в Выборге - 58,1 тыс. рублей. 1 кв.м. в Санкт-Петербурге стоит почти в два раза больше, чем в Выборге.
